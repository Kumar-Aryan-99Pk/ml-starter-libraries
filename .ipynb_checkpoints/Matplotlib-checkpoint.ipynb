# Matplotlib

to visualise data through graphs and charts

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x, y)
plt.show()
```

## Line PLot

```python
x = np.arange(0,10)
y = x**2

plt.plot(x, y)
plt.show()
```

## Scatter Plot

```python
x = [1,2,3,4,5]
y = [2,4,5,4,5]

plt.scatter(x, y)
plt.show()
```

## Bar Chart

```python
labels = ["A","B","C"]
values = [10,20,15]

plt.bar(labels, values)
plt.show()
```

## Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.show()
```

## Boxplot

```python
data = np.random.randn(100)

plt.boxplot(data)
plt.show()
```

## Customization

```python
plt.plot(x, y)
plt.title("Graph Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=np.array([1,2,3,5,6,8])
y=x**2
plt.plot(x, x, label="x")
plt.plot(x, y, label="x^2")
plt.legend()
plt.show()
```

```python
plt.plot(x, y, color="red", linestyle="--", marker="o")
```

+ color: "red", "blue", "green"
+ linestyle: "-", "--", ":"
+ marker: "o", "s", "^"

## Subplots

```python
x = np.arange(0,10)

plt.subplot(1,2,1)
plt.plot(x, x)

plt.subplot(1,2,2)
plt.plot(x, x**2)

plt.show()
```

```python
x=np.array([1,2,3,5,6,8])
y=x**2
plt.figure(figsize=(8,5))
plt.plot(x, y)
plt.show()
```

```python
#grids
plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y)
plt.savefig("plot.png")
```

```python
loss = [0.9, 0.7, 0.5, 0.3, 0.2]

plt.plot(loss)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
actual = [1,2,3,4,5]
pred = [1.1,1.9,3.2,3.8,5.1]

plt.scatter(actual, pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

## Heatmap

```python
data = np.random.rand(5,5)

plt.imshow(data, cmap="viridis")
plt.colorbar()
plt.show()
```
