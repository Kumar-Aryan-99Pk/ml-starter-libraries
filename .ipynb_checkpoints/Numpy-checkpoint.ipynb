# Numpy

Lib to manage arrays,matrices,statistics,etc

```python
import numpy as np
```

```python
#create array
a=np.array([1,4,5,67,80])
a
```

```python
#special arrays
z=np.zeros(3)
o=np.ones(4)
i=np.eye(3)
f=np.full((2,3),9)
l=np.arange(0,10,2)
print(z,o,"\n",i,"\n",f,l)
```

```python
#random
r=np.random.rand(2,3)
ri=np.random.randint(1,10,5) #low,high,size
print(r,"\n",ri)
```

```python
#array properties
print(f)
print(f.shape,"\n",f.ndim,"\n",f.size,"\n",f.dtype)
```

```python
#reshape
print(a)
print(a.reshape(1,5)) #row vector
print(a.reshape(5,1)) #col vector
print(r.flatten())
```

```python
#indexing

#for arrays
a[0]
a[0:3]

#for 2d arrays
r[1,2] #element at row 1,col 2
r[:,1] #all rows, col 1
r[1,:] #row 1, all col
```

## Important Math func

```python
np.sqrt(a)
np.exp(a)
np.log(a)
np.sin(a)
np.cos(a)
```

```python
#Matrix multiplication
np.dot(a, l)
a @ l
```

```python
#transpose
r.T
```

## Aggregate func

```python
np.sum(a)
np.mean(a)
np.median(a)
np.std(a)
np.var(a)
np.min(a)
np.max(a)
```

## Other

```python
np.sort(a)
np.unique(a)
np.isnan(a)
```
