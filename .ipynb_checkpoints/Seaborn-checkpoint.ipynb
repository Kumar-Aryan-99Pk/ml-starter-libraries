# Seaborn

advance data visualization lib built on Matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset("iris")
df.head()
```

# Scatter Plot

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=df)
plt.show()
```

# Line PLot

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=df)
plt.show()
```

# Bar Plot

```python
sns.barplot(x="species", y="sepal_length", data=df)
plt.show()
```

# Histogram

```python
sns.histplot(df["sepal_length"], bins=20)
plt.show()
```

# KDE Plot

```python
sns.kdeplot(df["sepal_length"])
plt.show()
```

# Box PLot

```python
sns.boxplot(x="species", y="sepal_length", data=df)
plt.show()
```

# Violin PLot

```python
sns.violinplot(x="species", y="sepal_length", data=df)
plt.show()
```

# Count Plot

```python
sns.countplot(x="species", data=df)
plt.show()
```

# Hue
Adds a third var using color

```python
sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    data=df
)
```

# Pair PLot

```python
sns.pairplot(df, hue="species")
```

# Heatmap

```python
corr = df.select_dtypes(include="number").corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

# Regression Plot

```python
sns.regplot(
    x="sepal_length",
    y="sepal_width",
    data=df
)
```

# Joint Plot

```python
sns.jointplot(
    x="sepal_length",
    y="sepal_width",
    data=df
)
```

# Styles

```python
sns.set_palette("muted")
sns.set_palette("pastel")
sns.set_palette("deep")
```
