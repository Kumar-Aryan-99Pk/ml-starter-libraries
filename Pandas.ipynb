To manage Series and Dataframes

````python
import pandas as pd
````

````python
#series => 1D
s=pd.Series([1,2,4,67,43])
s
````

````python
#dataframe => 2D
data={
    "Name":["A","Ary","Aryan"],
    "Age":[20,23,12]
}
df=pd.DataFrame(data)
df
````

## Viewing Data

````python
df.head()       # first 5 rows
df.tail()       # last 5 rows
df.shape        # (rows, columns)
df.columns      # column names
df.index        # row labels
df.info()       # data types
df.describe()   # statistics
````

## Selecting data

````python
df["Age"]
df[["Name", "Age"]]
````

````python
#by label
df.loc[0]
df.loc[0:2]
````

````python
#by pos
df.iloc[0]
df.iloc[0:3]  #notice the diff here
````

````python
#conditional selection
df[df["Age"] > 20]
````

## Modifying data

````python
#adding new col
df["Salary"] = [20000, 25000, 30000]
df
````

````python
df.drop("Age", axis=1)
````

````python
df.drop(0, axis=0)
````

<p>for permanent change use inplace=True</p>

````python
#missing values
df.isnull()
df.isnull().sum()
````

````python
#remove missing value
df.dropna()
````

#filling missing value
```
df.fillna(0)
df.fillna(df.mean())
```

````python
#sorting data
df.sort_values("Age")
df.sort_values("Age", ascending=False)
df.sort_values(["Age", "Salary"])
````

#grouping data
```
df.groupby("Department").mean()
df.groupby("Department")["Salary"].sum()
```

````python
#applying
df["Age"].apply(lambda x: x + 1)
````

````python
#value counts
df["Age"].value_counts()
df["Age"].unique()
df["Age"].nunique()
````

#correlation/stats
```
df.corr()
df["Age"].mean()
df["Salary"].std()
```
